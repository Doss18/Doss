```python
import pandas as pd
import numpy as np
import yfinance as yf

# Function to calculate the average momentum
def calculate_average_momentum(data, period=14):
    # Calculate momentum as the difference between today's close and the close 'period' days ago
    momentum = data['Close'].diff(period)
    avg_momentum = momentum.tail(period).mean()
    return avg_momentum

# Function to calculate average volatility (using standard deviation of returns)
def calculate_average_volatility(data, period=14):
    # Calculate daily returns
    returns = data['Close'].pct_change()
    volatility = returns.rolling(window=period).std().mean()
    return volatility

# Function to calculate price action average (average of close price)
def calculate_price_action_average(data, period=14):
    price_action_avg = data['Close'].rolling(window=period).mean().mean()
    return price_action_avg

# Fetching historical data using Yahoo Finance
def fetch_data(ticker, start_date, end_date):
    data = yf.download(ticker, start=start_date, end=end_date)
    return data

# Main function to call the calculations
def main():
    ticker = 'AAPL'  # Example: Apple stock
    start_date = '2023-01-01'
    end_date = '2024-01-01'

    # Fetch data
    data = fetch_data(ticker, start_date, end_date)

    # Calculate average momentum, volatility, and price action
    avg_momentum = calculate_average_momentum(data)
    avg_volatility = calculate_average_volatility(data)
    avg_price_action = calculate_price_action_average(data)

    # Print results
    print(f"Average Momentum: {avg_momentum}")
    print(f"Average Volatility: {avg_volatility}")
    print(f"Average Price Action: {avg_price_action}")

if __name__ == "__main__":
    main()
```

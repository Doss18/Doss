```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf

# Fetch historical data from Yahoo Finance
symbol = 'AAPL'  # Example: Apple Inc.
data = yf.download(symbol, start='2020-01-01', end='2025-01-01')

# Calculate the Momentum: Rate of change over the past 'n' days
n = 14  # Look-back period for momentum
data['Momentum'] = data['Close'].diff(n)

# Calculate Moving Averages for Price Action
short_window = 20  # Short-term moving average window
long_window = 50   # Long-term moving average window

data['Short_MA'] = data['Close'].rolling(window=short_window, min_periods=1).mean()
data['Long_MA'] = data['Close'].rolling(window=long_window, min_periods=1).mean()

# Strategy: Simple Price Action - Buy when Short MA crosses above Long MA, sell when opposite
data['Signal'] = 0
data['Signal'][short_window:] = np.where(data['Short_MA'][short_window:] > data['Long_MA'][short_window:], 1, 0)
data['Position'] = data['Signal'].diff()

# Plotting the data
plt.figure(figsize=(12,6))

# Price chart with Moving Averages
plt.subplot(2, 1, 1)
plt.plot(data['Close'], label='Close Price', color='blue')
plt.plot(data['Short_MA'], label=f'{short_window}-day Moving Average', color='red')
plt.plot(data['Long_MA'], label=f'{long_window}-day Moving Average', color='green')
plt.title(f'{symbol} Price and Moving Averages')
plt.legend()

# Momentum chart
plt.subplot(2, 1, 2)
plt
```

```python
# Importing necessary libraries
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Hypothetical dataset (this would be replaced by real weather data)
data = {
    'Humidity': [85, 90, 80, 70, 75, 78, 88, 91, 86, 92],  # in percentage
    'Wind Speed': [15, 18, 20, 12, 14, 16, 11, 19, 22, 13],  # in km/h
    'Temperature': [34, 33, 35, 32, 33, 32, 34, 33, 32, 34]  # in Celsius
}

# Creating a DataFrame from the data
df = pd.DataFrame(data)

# Features (Humidity and Wind Speed) and Target (Temperature)
X = df[['Humidity', 'Wind Speed']]
y = df['Temperature']

# Initializing the linear regression model
model = LinearRegression()

# Fitting the model to the data
model.fit(X, y)

# Predicting temperature based on new input data (for Tamil Nadu)
new_data = np.array([[85, 15]])  # Example: Humidity = 85%, Wind Speed = 15 km/h
predicted_temperature = model.predict(new_data)

# Displaying the predicted temperature
print(f"Predicted Temperature in Tamil Nadu: {predicted_temperature[0]:.2f}Â°C")

# Optional: Plotting the relationship
plt.scatter(df['Humidity'], df['Temperature'], color='blue', label='Humidity vs Temperature')
plt.scatter(df['Wind Speed'], df['Temperature'], color='red', label='Wind Speed vs Temperature')
plt.xlabel('Humidity/Wind Speed')
plt.ylabel('Temperature')
plt.title('Weather Data Analysis')
plt.legend()
plt.show()
```

```python
# Importing necessary libraries
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Hypothetical stock data (dates and closing prices in INR)
data = {
    'Date': pd.date_range(start="2025-03-01", periods=10, freq='D'),
    'Close Price': [1500, 1520, 1535, 1550, 1570, 1590, 1605, 1620, 1640, 1650]  # Stock closing price in INR
}

# Creating a DataFrame from the data
df = pd.DataFrame(data)

# We use the date index as a feature (convert to ordinal for prediction)
df['Date_Ordinal'] = df['Date'].apply(lambda x: x.toordinal())

# Features (Date_Ordinal) and Target (Close Price)
X = df[['Date_Ordinal']]
y = df['Close Price']

# Initializing the linear regression model
model = LinearRegression()

# Fitting the model to the data
model.fit(X, y)

# Predicting stock prices for the next two days
last_date_ordinal = df['Date_Ordinal'].iloc[-1]
next_day_ordinal_1 = last_date_ordinal + 1
next_day_ordinal_2 = last_date_ordinal + 2

# Predicting prices for the next 2 days
predicted_price_1 = model.predict(np.array([[next_day_ordinal_1]]))
predicted_price_2 = model.predict(np.array([[next_day_ordinal_2]]))

# Displaying the predictions for the next 2 days
next_day_1 = df['Date'].iloc[-1] + pd.Timedelta(days=1)
next_day_2 = df['Date'].iloc[-1] + pd.Timedelta(days=2)

print(f"Predicted stock price on {next_day_1.strftime('%Y-%m-%d')}: â¹{predicted_price_1[0]:.2f}")
print(f"Predicted stock price on {next_day_2.strftime('%Y-%m-%d')}: â¹{predicted_price_2[0]:.2f}")

# Optional: Plotting the historical data and prediction
plt.plot(df['Date'], df['Close Price'], color='blue', label='Historical Stock Prices')
plt.scatter([next_day_1, next_day_2], [predicted_price_1[0], predicted_price_2[0]], color='red', label='Predicted Prices')
plt.xlabel('Date')
plt.ylabel('Stock Price (INR)')
plt.title('Stock Price Prediction for Next 2 Days')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
